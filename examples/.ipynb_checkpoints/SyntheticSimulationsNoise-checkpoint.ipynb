# Set-Up

```julia
# Arvind Prasadan
# 2019 May
# DMF Package 
# Simulations

# DMF Package
push!(LOAD_PATH, "../src/")
using DMF

# Basic Linear Algebra Functionality
using LinearAlgebra
using Statistics
using StatsBase

# Plotting and Output
using Plots
using Measures
using LaTeXStrings
using Interact

# File IO
using FileIO
using JLD2

# Set plotting interface
pyplot()
```

## Parameters

```julia
n_list = Int.(round.(exp10.(range(log10(500.0); stop = 4.0, length = 10))))

k = 2 # Number of modes

# Mode vectors
p = 500

# Noise Variance
sigma_list = [0.0; exp10.(range(-2.0; stop = log10(2.0), length = 15))]
sigma_list = [sigma_list; 10.0.^(log10(2.0) .+ diff(log10.(sigma_list))[end] * (1:1:4))]

# Cosine Parameters
f_first = [0.25; 0.5]
f_second = [0.25; 2.0]
fs = 1.0

# ARMA Parameters
arma_std = 1.0
ar = [[0.2; 0.7], [0.3; 0.5]]
ma = [[], []]

trials = 100 # For missing data, noise, ARMA

trace_size = (400, 250); # For plotting
```

# Cosines

```julia
# Eigenvectors
Q = randn(p, k)
Q = mapslices(normalize, Q; dims = 1)

# Two time series
S_full = [gen_cos_sequence(n_list[end], f_second[1], fs)[1] gen_cos_sequence(n_list[end], f_second[2], fs)[1]]

# Magnitudes
D = Diagonal([2.0; 1.0])

# True Eigenvalues
w_true = sort(cos.(f_second))

# Results
cos_eigenvalue_error_f2 = zeros(length(n_list), length(sigma_list), trials, k, 3)
cos_eigenvector_error_f2 = zeros(length(n_list), length(sigma_list), trials, 3)
cos_signal_error_f2 = zeros(length(n_list), length(sigma_list), trials, 3)

t1 = time()
for nn = 1:1:length(n_list) # Loop over signal length
    n = n_list[nn]
    
    S = mapslices(normalize, S_full[1:n, :]; dims = 1)
    X = Q * D * S'
    
    for ss = 1:1:length(sigma_list) # Loop over noise variance
        println((time() - t1, nn / length(n_list), ss / length(sigma_list)))
        
        sigma = sigma_list[ss]
        
        for tr = 1:1:trials
            Xn = X + (sigma / sqrt(n)) * randn(p, n)

            w1, Q1, C1, _ = dmf(Xn; C_nsv = k, lag = 1)
            w2, Q2, C2, _ = dmf(Xn; tsvd = true, nsv = k, C_nsv = k, lag = 1)
            w3, Q3, C3, _ = dmf(Xn; os = true, nsv = k, C_nsv = k, lag = 1)

            # Compute (squared) error
            cos_eigenvalue_error_f2[nn, ss, tr, :, 1] = (w_true - sort(real.(w1[1:k]))[:]).^2.0
            cos_eigenvalue_error_f2[nn, ss, tr, :, 2] = (w_true - sort(real.(w2[1:k]))[:]).^2.0
            cos_eigenvalue_error_f2[nn, ss, tr, :, 3] = (w_true - sort(real.(w3[1:k]))[:]).^2.0

            cos_eigenvector_error_f2[nn, ss, tr, 1] = eigenvector_error(Q, Q1[:, 1:k])
            cos_eigenvector_error_f2[nn, ss, tr, 2] = eigenvector_error(Q, Q2[:, 1:k])
            cos_eigenvector_error_f2[nn, ss, tr, 3] = eigenvector_error(Q, Q3[:, 1:k])

            cos_signal_error_f2[nn, ss, tr, 1] = eigenvector_error(S, C1)
            cos_signal_error_f2[nn, ss, tr, 2] = eigenvector_error(S, C2)
            cos_signal_error_f2[nn, ss, tr, 3] = eigenvector_error(S, C3)
        end
    end
end
t2 = time()
println("$(t2 - t1) seconds")
```

# ARMA

```julia
# Eigenvectors
# Q = randn(p, k)
# Q = mapslices(normalize, Q; dims = 1)

# Magnitudes
# D = Diagonal([2.0; 1.0])

# True Eigenvalues: Lag 1, Lag 2
w1_true = sort([0.6; 2.0 / 3.0])
w2_true = sort([0.68; 5.0 / 6.0])

# Results

arma_eigenvalue_error_l1 = zeros(length(n_list), length(sigma_list), trials, k, 3)
arma_eigenvector_error_l1 = zeros(length(n_list), length(sigma_list), trials, 3)
arma_signal_error_l1 = zeros(length(n_list), length(sigma_list), trials, 3)

arma_eigenvalue_error_l2 = zeros(length(n_list), length(sigma_list), trials, k, 3)
arma_eigenvector_error_l2 = zeros(length(n_list), length(sigma_list), trials, 3)
arma_signal_error_l2 = zeros(length(n_list), length(sigma_list), trials, 3)

t1 = time()
for tr = 1:1:trials
    S_full = [gen_arma_sequence(n_list[end], ar[1], ma[1], arma_std) gen_arma_sequence(n_list[end], ar[2], ma[2], arma_std)] 
    
    for ss = 1:1:length(sigma_list) # Loop over noise variance
        println((time() - t1, tr / trials, ss / length(sigma_list)))
        
        sigma = sigma_list[ss]
        
        for nn = 1:1:length(n_list) # Loop over signal length 
            n = n_list[nn]
            
            S = mapslices(normalize, S_full[1:n, :]; dims = 1)
            X = Q * D * S'
            
            Xn = X + (sigma / sqrt(n)) * randn(p, n)

            # Lag 1
            
            w1, Q1, C1, _ = dmf(Xn; C_nsv = k, lag = 1)
            w2, Q2, C2, _ = dmf(Xn; tsvd = true, nsv = k, C_nsv = k, lag = 1)
            w3, Q3, C3, _ = dmf(Xn; os = true, nsv = k, C_nsv = k, lag = 1)

            # Compute (squared) error
            arma_eigenvalue_error_l1[nn, ss, tr, :, 1] = (w1_true - sort(real.(w1[1:k]))[:]).^2.0
            arma_eigenvalue_error_l1[nn, ss, tr, :, 2] = (w1_true - sort(real.(w2[1:k]))[:]).^2.0
            arma_eigenvalue_error_l1[nn, ss, tr, :, 3] = (w1_true - sort(real.(w3[1:k]))[:]).^2.0

            arma_eigenvector_error_l1[nn, ss, tr, 1] = eigenvector_error(Q, Q1[:, 1:k])
            arma_eigenvector_error_l1[nn, ss, tr, 2] = eigenvector_error(Q, Q2[:, 1:k])
            arma_eigenvector_error_l1[nn, ss, tr, 3] = eigenvector_error(Q, Q3[:, 1:k])

            arma_signal_error_l1[nn, ss, tr, 1] = eigenvector_error(S, C1)
            arma_signal_error_l1[nn, ss, tr, 2] = eigenvector_error(S, C2)
            arma_signal_error_l1[nn, ss, tr, 3] = eigenvector_error(S, C1)
            
            # Lag 2
            
            w1, Q1, C1, _ = dmf(Xn; C_nsv = k, lag = 2)
            w2, Q2, C2, _ = dmf(Xn; tsvd = true, nsv = k, C_nsv = k, lag = 2)
            w3, Q3, C3, _ = dmf(Xn; os = true, nsv = k, C_nsv = k, lag = 2)

            # Compute (squared) error
            arma_eigenvalue_error_l2[nn, ss, tr, :, 1] = (w2_true - sort(real.(w1[1:k]))[:]).^2.0
            arma_eigenvalue_error_l2[nn, ss, tr, :, 2] = (w2_true - sort(real.(w2[1:k]))[:]).^2.0
            arma_eigenvalue_error_l2[nn, ss, tr, :, 3] = (w2_true - sort(real.(w3[1:k]))[:]).^2.0

            arma_eigenvector_error_l2[nn, ss, tr, 1] = eigenvector_error(Q, Q1[:, 1:k])
            arma_eigenvector_error_l2[nn, ss, tr, 2] = eigenvector_error(Q, Q2[:, 1:k])
            arma_eigenvector_error_l2[nn, ss, tr, 3] = eigenvector_error(Q, Q3[:, 1:k])

            arma_signal_error_l2[nn, ss, tr, 1] = eigenvector_error(S, C1)
            arma_signal_error_l2[nn, ss, tr, 2] = eigenvector_error(S, C2)
            arma_signal_error_l2[nn, ss, tr, 3] = eigenvector_error(S, C3)
        end
    end
end
t2 = time()
println("$(t2 - t1) seconds")
```
